```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl

df = pd.read_excel('part a data sheet.xlsx', sheet_name='part a video 1')
data = df.to_numpy()


t= df.iloc[:,0]
vy = df.iloc[:,2]


region=(t>=0.01) & (t<=0.57)
t_region=t[region]
velocity_region=vy[region]

fit=np.polyfit(t_region, velocity_region,1)
bestfitline=np.polyval(fit, t_region)

slope=fit[0]
intercept=fit[1]

print(f"Best fit line equation: y = {slope:.2f}x + {intercept:.2f}")


plt.plot(t_region, bestfitline,color='red',ls='--')
plt.scatter(t, vy)
plt.xlabel('t')
plt.ylabel('vel')
plt.title('video 1 Velocity vs t')
plt.show()


print(f"therefore terminal velocity={slope:.2f}")
```
