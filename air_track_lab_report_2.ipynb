Part A plots, proving newtons first law, rectilinear motion

```python
import numpy as np
import matplotlib.pyplot as plt

d= [0.5, 1.0, 1.5, 2.0, 2.5]


v= [11.686, 19.995, 22.523, 34.867, 50.375]


plt.scatter(d,v)

coeff = np.polyfit( d, v, 1)

best_fit_line = np.polyval(coeff, d)

plt.plot(d, best_fit_line, linestyle='dashed',  color='blue', label='Best fit line ')

#print(f"best fit line equation: {coeff[0]:.4f}*y + {coeff[1]:.4f}")


yerr = 4.2 # average error percentage

# Plot with error bars
plt.errorbar(d, v, yerr=yerr, fmt='|', capsize=4, label='Data with Error Bars')


#plt.plot(v,d)
plt.xlabel('displacement(m)')
plt.ylabel('velocity(m/s)')
plt.grid(True)
plt.legend()
plt.title('Rectilinear motion')
plt.show()

```

Part C inelastic collisions


```python
import numpy as np
import matplotlib.pyplot as plt

# Length of the track (constant for all trials)
length_b = np.array([8.0] * 7)

# Time arrays for initial, final velocities and acceleration (in seconds)
time_i = np.array([0.16761, 0.14260, 0.12095, 0.09821, 0.09285, 0.07685, 0.07050])
time_f = np.array([0.138957, 0.11436, 0.10124, 0.08195, 0.07652, 0.06104, 0.05594])
time_acc = np.array([0.36938, 0.30918, 0.26690, 0.21772, 0.20553, 0.15997, 0.15036])

# Mass array in grams (since error is 0.1 grams)
mass_b = np.array([5, 8, 10, 15.2, 17.4, 32.5, 43.4])

# Error in mass (in grams)
mass_error = np.full_like(mass_b, 0.1)

# Calculating initial and final velocities in m/s
velocity_i = length_b / time_i
velocity_f = length_b / time_f

# Calculating acceleration in m/s²
acceleration = (velocity_f - velocity_i) / time_acc

# Scatter plot for Acceleration vs. Mass (Mass on y-axis)
plt.figure(figsize=(8, 6))
plt.errorbar(acceleration, mass_b, yerr=mass_error, fmt='o', color='blue', ecolor='slateblue', capsize=4, markersize=4, label='Data Points with Error Bars')

# Best-fit line calculation (linear regression)
coeff = np.polyfit(acceleration, mass_b, 1)  # Fits a 1st-degree polynomial (linear)
best_fit_line = np.polyval(coeff, acceleration)

# Plot best-fit line
plt.plot(acceleration, best_fit_line, color='indigo', linestyle='--', label=f'Best Fit Line: y={coeff[0]:.4f}x + {coeff[1]:.4f}')

print(acceleration)

# Labels and title
plt.ylabel('Acceleration (cm/s²)')
plt.xlabel('Mass (g)')
plt.title('Acceleration vs mass')
plt.grid(True)
plt.legend()

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Displacement in cm, starting from origin
d = [0, 0.5, 1.0, 1.5, 2.0, 2.5]

# Velocity in cm/s, starting from origin
v = [0, 11.686, 19.995, 22.523, 34.867, 50.375]

# Scatter plot of data, with indigo color for data points
plt.scatter(d, v, color='red', label='Data Points')

# Best fit line calculation
coeff = np.polyfit(d, v, 1)
best_fit_line = np.polyval(coeff, d)

# Plot best fit line, with purple color
plt.plot(d, best_fit_line, linestyle='dashed', color='blue', label='Best Fit Line')

# Error for displacement in cm
xerr = 0.1  # Error in centimeters

# Error for velocity in cm/s (already provided as 4.2 average error)
yerr = 4.2  # Error in cm/s

# Plot with red error bars
plt.errorbar(d, v, yerr=yerr, xerr=xerr, fmt='|', ecolor='slateblue', capsize=4, label='Error Bars')

# Label axes and add grid, legend, and title
plt.xlabel('Displacement (cm)')
plt.ylabel('Velocity (cm/s)')
plt.grid(True)
plt.legend()
plt.title('Rectilinear Motion')

# Show plot
plt.show()
```
