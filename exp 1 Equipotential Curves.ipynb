Verifying the formula for Part A: V(y)= V * (y/d)

```python
import matplotlib.pyplot as plt
import numpy as np
Vo=4.79
d= 10
y=[8,6.5, 5, 3.5, 2]

V_exp=[3.70, 3.07, 2.36, 1.67, 1.02] # Values of voltage while doing the experiment

V_form= [] #obtained values of voltage while verifying the formula

for val in y:
  V=Vo*(val/d)
  V_form.append(V)

#print(f"Expected Values of Voltage:{V_exp}")



#plt.scatter(y, V_form ,label= 'Formula Values') #x axis= distance from lower rod (y), y axis= voltage obtained through formula (V_form)
plt.scatter(y,V_exp, label= 'Experiment Values') #x axis= distance from lower rod (y), y axis= voltage from experiment (V_exp)

coeff = np.polyfit( y, V_exp, 1)

best_fit_line = np.polyval(coeff, y)

plt.plot(y, best_fit_line, linestyle='dashed',  color='blue', label='Best fit line ')

print(f"best fit line equation: {coeff[0]:.4f}*y + {coeff[1]:.4f}")


plt.xlabel('Distance from lower bar (cm)')
plt.ylabel('Voltage (V)')


plt.title('Part A: Variation of Voltage with Distance')
plt.grid(True)
plt.legend()
plt.show()
```

Veryifying the formula for Part B: V(r)= ((V1-V) ln(r/ro))ln(r/ro)  + Vo

```python
import numpy as np
import matplotlib.pyplot as plt


ro=2  #Inner circle radius
Vo= 4.7 #Voltage at inner circle/ voltage between the electrodes (concentric circles)

r=[3, 4.1, 5, 6, 7]

V_of_R =[3.52, 2.55, 1.91, 1.35, 0.95] #expected vlaues of voltage while conducting the experiment
ln_rbyro= []#obtained values of voltage while verifying the formula


for r_val in r:
  V = np.log(r_val/ro)
  ln_rbyro.append(V)


plt.scatter(ln_rbyro, V_of_R, label ='experimental values')


coeff2 = np.polyfit(ln_rbyro, V_of_R , 1)

best_fit_line2 = np.polyval(coeff2, ln_rbyro)

plt.plot(ln_rbyro, best_fit_line2, linestyle='dashed',  color='blue', label='Best fit line ')


print(f"best fit line equation: {coeff2[0]:.4f}*y + {coeff2[1]:.4f}")


plt.xlabel('ln(r/ro)')
plt.ylabel('Voltage (V) ')

plt.title('Part B: Variation of Voltage with Distance')
plt.grid(True)
plt.legend()
plt.show()


# plotting V_at_R vs ln( (r/ro) or r)
```
