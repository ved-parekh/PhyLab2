```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV
df = pd.read_csv(r'C:\Users\Shagnik\Downloads\Phy Lab 2\PA_252.3 gm', skiprows=1)

# Extract columns for time and vy (assuming column order: t, position, vy)
t = df.iloc[:, 0]     # Time
vy = df.iloc[:, 2]    # Velocity (vy)

# Define the region of interest
time_start = 1   # Start time for the region of interest
time_end = 9     # End time for the region of interest

region = (t >= time_start) & (t <= time_end)
t_region = t[region]
vy_region = vy[region]

# Check if there are data points in the selected region
if t_region.empty or vy_region.empty:
    print(f"No data points found in the specified time range ({time_start} to {time_end} seconds).")
else:
    # Outlier removal: filter out points that are beyond 2 standard deviations from the mean
    vy_mean = vy_region.mean()
    vy_std = vy_region.std()
    filtered_region = (vy_region >= (vy_mean - 2 * vy_std)) & (vy_region <= (vy_mean + 2 * vy_std))
    t_filtered = t_region[filtered_region]
    vy_filtered = vy_region[filtered_region]

    # Fit a line to the filtered vy vs t data in the selected region
    fit = np.polyfit(t_filtered, vy_filtered, 1)
    bestfitline = np.polyval(fit, t_filtered)

    # Extract slope and intercept for the best fit line
    slope = fit[0]
    intercept = fit[1]

    # Display the best fit line equation
    print(f"Best fit line equation: vy = {slope:.2f}t + {intercept:.2f}")

    # Set y-axis limit for better visualization (adjust as needed)
    plt.ylim(-2.5, 0)

    # Plot the vy data and best fit line within the region of interest
    plt.scatter(t_filtered, vy_filtered, color='indigo', label='Velocity Data (vy)', s=3)
    plt.plot(t_filtered, bestfitline, color='red', label='Best Fit Line')
    plt.xlabel('Time (s)')
    plt.ylabel('Velocity (vy) (cm/s)')
    plt.title('Velocity vs Time (252.3 gm)')
    plt.legend()
    plt.show()

    # Print slope, which represents acceleration in cm/s²
    print(f"Therefore, Acceleration = {slope:.2f} cm/s²")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV
df = pd.read_csv(r'C:\Users\Shagnik\Downloads\Phy Lab 2\PA_303.1 gm', skiprows=1)

# Extract columns for time and vy (assuming column order: t, position, vy)
t = df.iloc[:, 0]     # Time
vy = df.iloc[:, 2]    # Velocity (vy)

# Define the region of interest
time_start = 1   # Start time for the region of interest
time_end = 9     # End time for the region of interest

region = (t >= time_start) & (t <= time_end)
t_region = t[region]
vy_region = vy[region]

# Check if there are data points in the selected region
if t_region.empty or vy_region.empty:
    print(f"No data points found in the specified time range ({time_start} to {time_end} seconds).")
else:
    # Outlier detection using IQR
    Q1 = vy_region.quantile(0.25)
    Q3 = vy_region.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter out the outliers
    non_outliers = (vy_region >= lower_bound) & (vy_region <= upper_bound)
    t_region_filtered = t_region[non_outliers]
    vy_region_filtered = vy_region[non_outliers]

    # Fit a line to vy vs t in the filtered region
    fit = np.polyfit(t_region_filtered, vy_region_filtered, 1)
    bestfitline = np.polyval(fit, t_region_filtered)

    # Extract slope and intercept for the best fit line
    slope = fit[0]
    intercept = fit[1]

    # Display the best fit line equation
    print(f"Best fit line equation: vy = {slope:.2f}t + {intercept:.2f}")

    # Set y-axis limits based on the range of vy values with a margin
    vy_min, vy_max = vy_region_filtered.min(), vy_region_filtered.max()
    plt.ylim(vy_min - 0.2, vy_max + 0.2)

    # Plot the vy data and best fit line within the region of interest
    plt.scatter(t_region_filtered, vy_region_filtered, color='indigo', label='Velocity Data (vy)', s=3)
    plt.plot(t_region_filtered, bestfitline, color='red', label='Best Fit Line')
    plt.xlabel('Time (s)')
    plt.ylabel('Velocity (vy) (cm/s)')
    plt.title('Velocity vs Time (303.1 gm)')
    plt.legend()
    plt.show()

    # Print slope, which represents acceleration in cm/s²
    print(f"Therefore, Acceleration = {slope:.2f} cm/s²")
```
