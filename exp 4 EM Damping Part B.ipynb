```python
#Part B 303.1 trial 1 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl


df = pd.read_excel('part b 303.1 trial 1.xlsx', sheet_name='part b 303.1 trial 1')
data = df.to_numpy()


t= df.iloc[:,0]
position= df.iloc[:,1]
vy = df.iloc[:,2]

region=(t>=9) & (t<=12)
t_region=t[region]
position_region=position[region]

fit=np.polyfit(t_region,position_region,1)
bestfitline=np.polyval(fit, t_region)

slope=fit[0]
intercept=fit[1]

print(f"Best fit line equation: y = {slope:.2f}x + {intercept:.2f}")

plt.scatter(t,position, color='blue')
plt.plot(t_region, bestfitline,color='red',linewidth=5)

plt.xlabel('t')
plt.ylabel('Position')
plt.title('Position vs t')
plt.show()


print(f"therefore terminal velocity={slope:.2f}")
```

```python
# plot for velocities vs mass, the slope of this line will give us the (gr^2/CB^2) the intercept of this equation would be (-rTf/CB^2)

import numpy as np
import matplotlib.pyplot as plt

m = [0.101, 0.152, 0.202, 0.253, 0.303, 0.353]  #mass in kg, 51.3 not found yet
v= [0.0285, 0.0376, 0.0528, 0.0689, 0.0826, 0.0978] # velocity in m/s, velocity for 51.3 notfoundyet 


fit=np.polyfit(m,v,1)
bestfitline=np.polyval(fit, m)

slope=fit[0]
intercept=fit[1]



print(f"Best fit line equation: y = {slope:.4f}x {intercept:.4f}")


plt.plot(m, bestfitline, color='red', label=f'Best Fit Line: y = {fit[0]:.4f}x {fit[1]:.4f}')
plt.scatter(m,v, color='blue', label='Data Points', s=50)
plt.xlabel('mass(kg)')
plt.ylabel('Velocity (m/s)')
plt.title('Velocity vs Mass')

plt.legend()
plt.show()
```
