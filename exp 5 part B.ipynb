```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl
```

```python
df = pd.read_excel('part b data new.xlsx', sheet_name='4 mags')
data = df.to_numpy()


t= df.iloc[:,0]
y =df.iloc[:, 1]
vy = df.iloc[:,2]


region1=(t>=0.55) & (t<=0.78)
t_region1=t[region1]
y_region1=y[region1]

m1,c1 =np.polyfit(t_region1, y_region1,1)

print(f"Best fit line equation: y = {m1:.2f}x + {c1:.2f}")

plt.scatter(t,y)

plt.plot(t_region1, m1*t_region1 + c1,color='red',ls='--')

plt.xlabel('t')
plt.ylabel('Position')
plt.title('4 mags 2210 Position vs t')
plt.show()


```

```python
yaxis= -2.85 * np.arctanh(vy/(-2.85))
plt.scatter(t, yaxis)
plt.xlabel('t') 
plt.ylabel('vt*arctanh(vy/vt)')

region2= (t>=0.02)&(t<=0.35)

t_region2= t[region2]
y_region2=yaxis[region2]

m2,c2 =np.polyfit(t_region2, y_region2, 1)
bestline= m2*t_region2+c2

print(f"Best fit line equation: y = {m2:.2f}x + {c2:.2f}")
plt.plot(t_region2, bestline , color='red',ls='--')

plt.title('4 mags, vt*arctanh(vy/vt) vs t')
```
